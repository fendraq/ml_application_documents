```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import ConfusionMatrixDisplay

cm = np.array([[8080, 197], [18, 1297]])
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["0", "1"])

fig, ax = plt.subplots(figsize=(11, 4))
im = ax.imshow(cm, interpolation='nearest', cmap='Blues')
disp.plot(ax=ax, cmap="Blues", colorbar=False)

# Lägg till färgbar till höger
cbar = fig.colorbar(im, ax=ax, fraction=0.03, pad=0.04)

plt.title("Confusion Matrix for Ensemble")
plt.xlabel("Predicted label")
plt.ylabel("True label")
plt.tight_layout()
plt.show()

offset_y = -0.15  # Justera detta värde efter behov

""" labels = np.array([["TN", "FP"], ["FN", "TP"]])
for (i, j), label in np.ndenumerate(labels):
    ax.text(
        j, i + offset_y,  # Flytta texten lite nedåt
        label,
        ha="center", va="center",
        color="black", fontsize=14, fontweight="bold"
    ) """

""" plt.figtext(
    0.65, 0.5,
    "TN: *True Negative*\n – Modellen förutsäger negativt och det är korrekt.\n"
    "FP: *False Positive*\n – Modellen förutsäger positivt men det är fel.\n"
    "FN: *False Negative*\n – Modellen förutsäger negativt men det är fel.\n"
    "TP: *True Positive*\n – Modellen förutsäger positivt och det är korrekt.",
    wrap=True, ha="left", va="center", fontsize=12, style="italic"
) """
```
